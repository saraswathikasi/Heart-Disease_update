```python
y = df["target"] #to have target column only
X = df.drop("target",axis=1) #here axis = 0 for rows it will look for target in rows where axis = 1 represents column so it chooses column target and remove it.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) #split the data into 80:20 for training and testing
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(y_test.unique())
Counter(y_train)
```

```python
m2 = 'Naive Bayes'
nb = GaussianNB()
nb.fit(X_train,y_train)
nbpred = nb.predict(X_test)
nb_conf_matrix = confusion_matrix(y_test,nbpred)
nb_acc_score = accuracy_score(y_test,nbpred)
print("confusion matrix")
print(nb_conf_matrix)
print("\n")
print("Accuracy of the Naive Bayes model is:",nb_acc_score*100,'\n')
print(classification_report(y_test,nbpred))
```

```python
m1 = 'Logistic Regression'
lr = LogisticRegression()
model = lr.fit(X_train,y_train)
lr_predict = lr.predict(X_test)
lr_conf_matrix = confusion_matrix(y_test,lr_predict)
lr_acc_score = accuracy_score(y_test,lr_predict)
print("confusion matrix")
print(lr_conf_matrix)
print("\n")
print("Accuracy of the Logistic Regression model is:",lr_acc_score*100,'\n')
print(classification_report(y_test,lr_predict))
```

```python
m3 = 'Random Forest Classifier'
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
rf_predicted = rf.predict(X_test)
rf_conf_matrix = confusion_matrix(y_test,rf_predicted)
rf_acc_score = accuracy_score(y_test,rf_predicted)
print("confusion matrix")
print(rf_conf_matrix)
print("\n")
print("Accuracy of the Random Forest Classifier model is:",rf_acc_score*100,'\n')
print(classification_report(y_test,rf_predicted))
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

m4 = 'Extreme Gradient Boosting Classifier'
xgb = XGBClassifier(
    learning_rate=0.01,
    n_estimators=25,
    max_depth=15,
    gamma=0.6,
    subsample=0.52,
    colsample_bytree=0.6,
    reg_lambda=2,
    booster='dart',
    colsample_bylevel=0.6,
    colsample_bynode=0.5
)

xgb.fit(X_train, y_train)
xgb_predicted = xgb.predict(X_test)

xgb_conf_matrix = confusion_matrix(y_test, xgb_predicted)
xgb_acc_score = accuracy_score(y_test, xgb_predicted)

print("Confusion Matrix:")
print(xgb_conf_matrix)
print("\nAccuracy of the Extreme Gradient Boosting Classifier model is:", xgb_acc_score * 100, '\n')
print(classification_report(y_test, xgb_predicted))
```

```python
m5 = 'K-Nearest Neighbors Classifier'
knn = KNeighborsClassifier(n_neighbors= 10)
knn.fit(X_train,y_train)
knn_predicted = knn.predict(X_test)
knn_conf_matrix = confusion_matrix(y_test,knn_predicted)
knn_acc_score = accuracy_score(y_test,knn_predicted)
print("confusion matrix")
print(knn_conf_matrix)
print("\n")
print("Accuracy of the K-Nearest Neighbors Classifier model is:",knn_acc_score*100,'\n')
print(classification_report(y_test,knn_predicted))
```

```python
m6 = 'Decision Tree Classifier'
dt = DecisionTreeClassifier(criterion='entropy', random_state=0,max_depth = 6)
dt.fit(X_train,y_train)
dt_predicted = dt.predict(X_test)
dt_conf_matrix = confusion_matrix(y_test,dt_predicted)
dt_acc_score = accuracy_score(y_test,dt_predicted)
print("confusion matrix")
print(dt_conf_matrix)
print("\n")
print("Accuracy of the Decision Tree Classifier model is:",dt_acc_score*100,'\n')
print(classification_report(y_test,dt_predicted))
```

```python
m7 = 'Support Vector Classifier'
svc = SVC(kernel = 'rbf', C = 2)
svc.fit(X_train,y_train)
svc_predicted = svc.predict(X_test)
svc_conf_matrix = confusion_matrix(y_test,svc_predicted)
svc_acc_score = accuracy_score(y_test,svc_predicted)
print("confusion matrix")
print(svc_conf_matrix)
print("\n")
print("Accuracy of the Support Vector Classifier model is:",svc_acc_score*100,'\n')
print(classification_report(y_test,svc_predicted))
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import StackingClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

# Define base classifiers
nb = GaussianNB()
lr = LogisticRegression(max_iter=1000)
svc = SVC(probability=True)

# Create stacking classifier
stack_model = StackingClassifier(
    estimators=[('nb', nb), ('lr', lr), ('svc', svc)],
    final_estimator=LogisticRegression(),
    cv=5
)

# Fit the model
stack_model.fit(X_train, y_train)

# Predict and evaluate
scv_predicted = stack_model.predict(X_test)
scv_conf_matrix = confusion_matrix(y_test, scv_predicted)
scv_acc_score = accuracy_score(y_test, scv_predicted)

# Output
print("Confusion Matrix:")
print(scv_conf_matrix)
print("\nAccuracy of the Stacking Classifier model is:", scv_acc_score * 100, '%\n')
print(classification_report(y_test, scv_predicted))
```

```python
from sklearn.ensemble import VotingClassifier

# Create base models
svm = SVC(probability=True, kernel='rbf', random_state=42)
lr = LogisticRegression(max_iter=1000, random_state=42)
nb = GaussianNB()

# Voting Classifier (Soft Voting for better performance)
voting_clf = VotingClassifier(
    estimators=[('svm', svm), ('lr', lr), ('nb', nb)],
    voting='soft'
)

# Train on your data
voting_clf.fit(X_train, y_train)

# Predict and evaluate
y_pred_voting = voting_clf.predict(X_test)

print("Voting Classifier Accuracy:", accuracy_score(y_test, y_pred_voting) * 100)
print("\nClassification Report:\n", classification_report(y_test, y_pred_voting))
```

What This Will Give You

Three separate plots:

SVM: Which features most affect its decision.

Logistic Regression: Which features it weights the most.

Naïve Bayes: Which features influence probability the most.

Ability to compare how different models focus on different features.

```python
import shap
import numpy as np

np.random.seed(42)

# Sample data
X_test_sampled = shap.sample(X_test, 50)
X_train_sampled = shap.sample(X_train, 50)

def get_shap_values(explainer, X_sample):
    shap_values = explainer.shap_values(X_sample)
    # If binary classification with list output, take the second element (positive class)
    if isinstance(shap_values, list) and len(shap_values) == 2:
        return shap_values[1]
    # Otherwise return the array directly
    return shap_values

# ---------------------------
# 1. SHAP for Support Vector Machine (SVM)
# ---------------------------
print("\n🔍 SHAP Explanation for SVM Model")
svm.fit(X_train, y_train)
explainer_svm = shap.KernelExplainer(svm.predict_proba, X_train_sampled)
shap_values_svm = get_shap_values(explainer_svm, X_test_sampled)
shap.summary_plot(shap_values_svm, X_test_sampled, feature_names=X.columns)

# ---------------------------
# 2. SHAP for Logistic Regression
# ---------------------------
print("\n🔍 SHAP Explanation for Logistic Regression Model")
lr.fit(X_train, y_train)
explainer_lr = shap.KernelExplainer(lr.predict_proba, X_train_sampled)
shap_values_lr = get_shap_values(explainer_lr, X_test_sampled)
shap.summary_plot(shap_values_lr, X_test_sampled, feature_names=X.columns)

# ---------------------------
# 3. SHAP for Naïve Bayes
# ---------------------------
print("\n🔍 SHAP Explanation for Naïve Bayes Model")
nb.fit(X_train, y_train)
explainer_nb = shap.KernelExplainer(nb.predict_proba, X_train_sampled)
shap_values_nb = get_shap_values(explainer_nb, X_test_sampled)
shap.summary_plot(shap_values_nb, X_test_sampled, feature_names=X.columns)
```
